```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_breast_cancer
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge, Lasso			
from sklearn.metrics import r2_score
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
```

```python
df= load_breast_cancer()
X=pd.DataFrame(df.data)
y = df.target
ridge_regression = Ridge()
lasso_regression = Lasso()
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=42)
```

```python
parametes= {'alpha': [1, 2, 3, 4, 5, 6, 7, 8, 20 ,30 , 20 , 40 , 50, 60, 70, 80, 90] }
ridge_cv = GridSearchCV( ridge_regression, parametes, cv=5, scoring='neg_mean_squared_error')
ridge_cv.fit(X_train, y_train)
```

```python
ridge_pred = ridge_cv.predict(X_test)
print(ridge_cv.best_params_)
print(ridge_cv.best_score_)
```

```python
sns.histplot(ridge_pred-y_test, kind='kde')
sns.histplot(ridge_pred-y_test, kde=True)
```
